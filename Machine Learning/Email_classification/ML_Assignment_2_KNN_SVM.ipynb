```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
df=pd.read_csv('emails.csv')
```

```python
df.head() #used to display the first few rows of a DataFrame.
```

```python
df.columns
```

```python
df.isnull().sum() #returns the total count of missing values in each column.No missing values found below if 0.
```

```python
df.dropna(inplace = True)
```

```python
df.drop(['Email No.'],axis=1,inplace=True)
X = df.drop(['Prediction'],axis = 1) #X = Input features → the data you give to the model so it can learn patterns.
y = df['Prediction'] #Prediction column likely contains labels spam or not spam.
```

```python
from sklearn.preprocessing import scale #scale() function from Scikit-learn’s preprocessing module.
X = scale(X) #standardizes the features by removing the mean and scaling to unit variance.
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42) #30% of data → used for testing, 70% → for training
```

```python
from sklearn.neighbors import KNeighborsClassifier     #Classifies a new data point based on its nearest neighbors in the training data
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train) 
y_pred = knn.predict(X_test) #Predicts the labels for the test data based on the trained model.
```

```python
print("Prediction",y_pred)
```

```python
print("KNN accuracy = ",metrics.accuracy_score(y_test,y_pred)) #compares the predicted labels with the actual labels from the test set to calculate the accuracy of the model.
```

```python
print("Confusion matrix",metrics.confusion_matrix(y_test,y_pred)) 
```

```python
# cost C = 1 , It controls how much the model allows misclassifications during training
model = SVC(C = 1) 

# Train the SVM on training data
model.fit(X_train, y_train)

# this line uses the trained model to predict labels for the unseen test data.
y_pred = model.predict(X_test)
```

```python
metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
print("SVM accuracy = ",metrics.accuracy_score(y_test,y_pred)) #compares the predicted labels with the actual labels from the test set to calculate the accuracy of the model.
```
